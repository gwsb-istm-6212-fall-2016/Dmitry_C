I am going to find the most common types of violations found by the inspectors

```python
%%sql
SELECT "VIOLATION_DESC", COUNT(*) AS count
FROM fy2010
GROUP BY "VIOLATION_DESC"
ORDER BY count DESC;
```

As can be seen from the table, in the majority of cases inspectors leave additional comments. However, if they do find a problem it is more likely to be related to plumbing issues.

```python
%%sql
SELECT "VIOLATION_CODE"
FROM fy2010
WHERE "VIOLATION_DESC" = 'Plumbing: Improperly installed/maintained/supplied'
LIMIT 1;
```

The violation code for plumbing issues is 42

```python
%%sql
SELECT "FACILITY_NAME", "SITE_ADDRESS","CITY", "STATE", "ZIP"
FROM fy2010
WHERE "VIOLATION_CODE" = '42';
```

The list above contains facilities with plumbing issues and their locations
